```python
def state(stat):
    # Wait for the dropdown to load
    dropdown = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'states'))
    )

    # Click on the dropdown to open the options
    dropdown.click()

    # Locate the "Uttarakhand" option and click it
    uttrakhand_option = WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.XPATH, "//option[@value='Uttarakhand']"))
    )
    uttrakhand_option.click()
```

```python
def city(cit):
    # Wait for the dropdown to load
    dropdown = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'cities'))
    )

    # Click on the dropdown to open the options
    dropdown.click()

    # Locate the "Uttarakhand" option and click it
    dehradun_option = WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.XPATH, "//option[@value='Dehradun']"))
    )
    dehradun_option.click()
```

```python
def stations(station):
    # Wait for the dropdown to load
    dropdown = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'stations'))
    )

    # Click on the dropdown to open the options
    dropdown.click()

    # Locate the "Uttarakhand" option and click it
    dehradun_option = WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.XPATH, "//option[@value='site_5594']"))
    )
    dehradun_option.click()
```
